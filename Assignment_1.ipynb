### 5.VISUALIZATIONS

### a.Histogram

```python
data[['SL','PL']].plot.hist(rwidth=0.9)
plt.title('Distrubution of length of Iris plants')
plt.xlabel('Sepal & Petal Length')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
data[['SW','PW']].plot.hist(rwidth=0.9)
plt.title('Distribution of Width of Iris plants')
plt.xlabel('Sepal & Petal Width')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

### b.Scatter Plot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data['SL'],data['PL'],s=9,c='red')
plt.title('Sepal length V/S Petal length')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')

plt.subplot(2,1,2)
plt.scatter(data['SW'],data['PW'],s=10,c='blue')
plt.title('Sepal Width V/S Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')

plt.tight_layout()
```
